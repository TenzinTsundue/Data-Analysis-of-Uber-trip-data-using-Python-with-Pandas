```python
%pylab inline
import pandas as pd
import seaborn as sns
```

## Load CSV file into memory

```python
data = pd.read_csv('uber-raw-data-apr14.csv')
```

```python
data.head()
```

## Convert datatime and add some useful columns

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
data.head()
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.head()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['hour'] = data['Date/Time'].map(get_hour)

data.head()
```

## Analysis

### Analyze the DoM (Date of Month)

```python
hist(data.dom, bins = 30, rwidth = .8, range = (0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - uber - Apr 2014')
```

```python
# for k, rows in data.gruopby('dom')
# print((k, len(rows)))

def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
bar(range(1, 31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1, 31), by_date_sorted)
xticks(range(1, 31), by_date_sorted.index)
xlabel('dat of the month')
ylabel('frequency')
title('Frequency by DoM - uber - Apr 2014')
```

## Analuze the hour

```python
hist(data.hour, bins = 24, range = (.5, 24))
```

## Analyze the weeday

```python
hist(data.weekday, bins = 7, range= (-.5, 6.5), rwidth = .8, color = '#AA6666', alpha= .4)
xticks(range(7), 'Mon Tue Wed Fri Sat Sun'.split())
```

## Cross analysis (hour, dow)

```python
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()
```

```python
sns.heatmap(by_cross)
```
